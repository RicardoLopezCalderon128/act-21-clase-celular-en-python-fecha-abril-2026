```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# Instanciación
celular1 = Celular("Samsung", "Galaxy S23", 800)

# Crear objetos (instancias)
mi_cel = Celular("Apple", "iPhone 15", 1000)
otro_cel = Celular("Xiaomi", "Redmi Note 12", 300)
print("celular 1")

celular1.mostrar_info()
celular1.encender()
print("\ncelular 2")

otro_cel.mostrar_info()
otro_cel.encender()
# Acceder a métodos
mi_cel.mostrar_info()
print("programa terminado")
print("creador de software por Lopez ricardo NL = 29")
```
